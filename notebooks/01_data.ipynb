```python
!rm -rf main_passwd_db.txt
```

```python
import re

passwds = []
one_million = 1000000
with open('passwords_printable_sorted.txt', encoding='latin-1') as file:
    for line in file:
        line = line.strip()
        if line.isascii() and len(line) > 3 and len(line) < 51 and len(list(filter(lambda y: not (32 <= ord(y) <=126), set(line)))) == 0 and len(set(line)) > 2 and len(re.findall(r'[\w.+-]+@[\w-]+\.[\w.-]+', line)) == 0:
          passwds.append(line.rstrip())
          if len(passwds) == 10 * one_million:
            print("Processed 10 M records")
            with open("main_passwd_db.txt", "a") as outfile:
              outfile.write("\n".join(passwds))
            passwds = []
```

# Build char distribution

```python
num_pass_w_char = {}

with open('main_passwd_db.txt') as file:
    for line in file:
        line = line.strip()
        if line[0] not in num_pass_w_char:
            num_pass_w_char[line[0]] = 1
        else:
            num_pass_w_char[line[0]] += 1
```
